```python
#openCV可用格式
import matplotlib.pyplot as plt
from pydicom import dcmread
from pydicom.data import get_testdata_files

# 取得 Pydicom 附帶的 DICOM 測試影像路徑
filename =('../input/rsna-2022-cervical-spine-fracture-detection/train_images/1.2.826.0.1.3680043.10001/1.dcm')
#filename = get_testdata_files('../input/rsna-2022-cervical-spine-fracture-detection/train_images/1.2.826.0.1.3680043.10001/1.dcm')
#filename = get_testdata_files("1.dcm")

# 讀取 DICOM 檔案
ds = dcmread(filename)
ds_arr = ds.pixel_array
lens = ds_arr.shape[0]*ds_arr.shape[1]
arr_temp = np.reshape(ds_arr,(lens,))
max_val = max(arr_temp)
min_val = min(arr_temp)
img_arr = (ds_arr-min_val)/(max_val-min_val)
plt.imshow(img_arr,cmap = plt.cm.bone)

print("shape:",ds.pixel_array.shape)
print("type:",ds.pixel_array.dtype)



```
